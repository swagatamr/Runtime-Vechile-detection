```python
#loading librariries and necessary files
import cv2
import imutils
cascade_src = r'cars.xml'
car_cascade = cv2.CascadeClassifier(cascade_src)
#camera
cam=cv2.VideoCapture(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
'''
cascade_src = r'cars.xml'
car_cascade = cv2.CascadeClassifier(cascade_src)
'''
```

For still image

```python

img=cv2.imread(r"C:\Users\swagatam\Documents\1.jpg")
img=imutils.resize(img,width=300) #resize to 300
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #color to Grayscale
cars = car_cascade.detectMultiScale(gray, 1.1, 1) #coordinates of vehicle in a frame
for (x,y,w,h) in cars:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
plt.imshow( img)
#cv2.destroyAllWindows()
```

For runntime detection with one camera

```python

while True:
    detected = 0
    _,img=cam.read() #reading frame from camera
    img=imutils.resize(img,width=300) #resize to 300
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #color to Grayscale
    cars = car_cascade.detectMultiScale(gray, 1.1, 1) #coordinates of vehicle in a frame
    for (x,y,w,h) in cars:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
    cv2.imshow("Frame", img)
    b=str(len(cars))
    a= int(b)
    detected=a
    n=detected
    print("------------------------------------------------")
    print ("North: %d "%(n))
    if n>=2:
        print ("North More Traffic")
    else:
        print ("no traffic")
    if cv2.waitKey(33) == 27:
        break

cam.release()
cv2.destroyAllWindows()

```

```python

while True:
    detected1 = 0
    _,img1=cam1.read() #reading frame from camera
    img1=imutils.resize(img1,width=300) #resize to 300
    gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY) #color to Grayscale
    cars = car_cascade.detectMultiScale(gray, 1.1, 1) #coordinates of vehicle in a frame
    for (x,y,w,h) in cars:
        cv2.rectangle(img1,(x,y),(x+w,y+h),(0,0,255),2)
    cv2.imshow("Car2", img1)
    b1=str(len(cars))
    a1= int(b1)
    detected1=a1
    n1=detected1
    
    detected = 0
    _,img=cam.read() #reading frame from camera
    img=imutils.resize(img,width=300) #resize to 300
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #color to Grayscale
    cars = car_cascade.detectMultiScale(gray, 1.1, 1) #coordinates of vehicle in a frame
    for (x,y,w,h) in cars:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
    cv2.imshow("Car1", img)
    b=str(len(cars))
    a= int(b)
    detected=a
    n=detected

    print("------------------------------------------------")
    print ("North: %d "%(n1))
    if n1>=2:
        print ("North More Traffic")
    else:
        print ("no traffic")
    if cv2.waitKey(33) == 27:
        break
    print("------------------------------------------------")
    print ("South: %d "%(n))
    if n>=2:
        print ("South More Traffic")
    else:
        print ("no traffic")
    if cv2.waitKey(33) == 27:
        break
cam.release()    
cam1.release()
cv2.destroyAllWindows()
```
